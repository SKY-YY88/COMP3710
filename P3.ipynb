```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Define vertices of the initial triangle
vertices = torch.tensor([[0.0, 0.0], [1.0, 0.0], [0.5, np.sqrt(3)/2]], device=device)

# Initialize a point inside the triangle
point = torch.tensor([0.5, 0.5], device=device)

# Number of iterations
num_iterations = 50000

# Create an array to store the points
points = torch.empty((num_iterations, 2), device=device)

# Iteratively generate the points
for i in range(num_iterations):
    chosen_vertex = vertices[torch.randint(3, (1,), device=device)]
    point = 0.5 * (point + chosen_vertex)
    points[i] = point

# Plot the Sierpinski triangle
plt.figure(figsize=(8, 8))
plt.scatter(points[:, 0].cpu().numpy(), points[:, 1].cpu().numpy(), s=0.1)
plt.axis('equal')
plt.axis('off')
plt.show()
```
