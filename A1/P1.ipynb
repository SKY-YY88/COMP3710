```python
import torch
import numpy as np
import matplotlib.pyplot as plt
print("PyTorch Version", torch.__version__)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
#grid for computing image, it subdivides the space

X, Y = np.mgrid[-4.0:4.0:0.01, -4.0:4.0:0.01]
x = torch.tensor(X)
y = torch.tensor(Y)
```

```python
x = x.to(device)
y = y.to(device)
```

```python
z = torch.exp(-(x**2+y**2)/2.0)
plt.imshow(z.cpu().numpy())
plt.tight_layout()
plt.show
```

```python
z2 = torch.sin(x)
plt.imshow(z2.cpu().numpy())
plt.tight_layout()
plt.show
```

```python
z3 = z*z2
plt.imshow(z3.cpu().numpy())
plt.tight_layout()
plt.show
```
